```python
import pandas as pd
import string
import numpy as np
import ast
import random
from datetime import datetime

## Section 1: Impacted flight unique DEP_KEY
selected_DEP_KEY = 'ZZ20240428HYDAMD3723'  # Three flight solution case
# selected_DEP_KEY = 'ZZ20240503MAAAMD2840'   # One flight solution case

## Section 2: Passenger and Flight Scoring Functions
def calculate_passenger_score(_class, downline, pax, ssr):
    # Cabin mapping based on class
    cabin_mapping = {random.choice(['F', 'P']): 'FirstClass', random.choice(['C', 'J', 'Z']): 'BusinessClass',
                     random.choice(['R','S','T','H', 'Q', 'M']): 'PremiumEconomyClass',
                     random.choice(['A', 'Y', 'B','D','E','G','I','K','L','N','O','U','V','W','X']): 'EconomyClass'}

    # Derive the cabin based on the class
    cabin = next((key for key, values in cabin_mapping.items() if _class in values), None)
    if cabin is None:
        raise ValueError(f"No matching cabin found for class {_class}")

    # Assign scores to cabins
    cabin_score = {letter: 500 + (i * (1000 - 500) / (26 - 1)) for i, letter in enumerate(string.ascii_uppercase)}

    # Calculate class scores dynamically
    class_scores = {'FirstClass': 2700, 'BusinessClass': 2000, 'PremiumEconomyClass': 1500, 'EconomyClass': 1000}

    # Score calculations
    downline_score = 100 * downline
    pax_score = 50 * pax
    ssr_score = 200 * ssr

    total_score = cabin_score[cabin] + class_scores[_class] + downline_score + pax_score + ssr_score
    return cabin, cabin_score[cabin], class_scores[_class], downline_score, pax_score, ssr_score, total_score

def calculate_flight_score(arrival_delay, spf, equipment, CP_waiting_time_difference, CP_same_airport):
    # Score calculations
    arrival_delay_score = 70 if arrival_delay <= 6 else (50 if arrival_delay <= 12 else (40 if arrival_delay <= 24 else 30))
    spf_score = 70 if spf <= 6 else (60 if spf <= 12 else (50 if spf <= 24 else (40 if spf <= 48 else 30)))

    # Modify city_pairs_score based on additional conditions
    if CP_waiting_time_difference < 0.5:
        city_pairs_score = 50
    elif CP_same_airport:
        city_pairs_score = 40
    else:
        city_pairs_score = 60

    if equipment:
        equipment_score = 50
    else:
        equipment_score=0

    total_flight_score = arrival_delay_score + spf_score + city_pairs_score + equipment_score
    return arrival_delay_score, spf_score, city_pairs_score, equipment_score, total_flight_score

def assign_grade(total_score):
    # Assign grades based on the total score
    if total_score >= 200: return 'A'
    elif total_score >= 180: return 'B'
    elif total_score >= 150: return 'C'
    else: return 'D'

## Section 3: Initialize DataFrames for Scores and Grades
proposed_classes = {
    'A': ['F', 'S'],
    'S': ['F', 'A'],
    'F': ['S', 'A'],
    'J': ['O', 'C', 'I'],
    'C': ['O', 'J', 'I'],
    'I': ['J', 'C', 'O'],
    'O': ['J', 'C', 'I'],
    'Y': ['B', 'P', 'Z'],
    'B': ['Y', 'P', 'Z']
}

def accessible_inventory(inventory_df):
    inventory_df['FC_CD'] = (inventory_df['FC_CD']).apply(ast.literal_eval)
    inventory_df['BC_CD'] = (inventory_df['BC_CD']).apply(ast.literal_eval)
    inventory_df['PC_CD'] = (inventory_df['PC_CD']).apply(ast.literal_eval)
    inventory_df['EC_CD'] = (inventory_df['EC_CD']).apply(ast.literal_eval)
    return inventory_df

def allocate_seats(oversold, classes):
    allocated_seats = {}
    for cls, total_seats in classes.items():
        available_seats = min(total_seats, oversold)
        allocated_seats[cls] = available_seats
        oversold -= available_seats
    return allocated_seats

def distribute_equally(oversold, classes):
    class_counts = len(classes)
    seats_per_class = oversold // class_counts
    remaining_seats = oversold % class_counts

    available_seats_dict = {}
    for cls, total_seats in classes.items():
        allocated_seats = seats_per_class + (remaining_seats > 0)
        remaining_seats -= 1
        allocated_seats = min(allocated_seats, total_seats)

        available_seats_dict[cls] = allocated_seats
        classes[cls] -= allocated_seats

    return available_seats_dict

def calculate_score(exception_list_length):
    max_score = 100
    decrease_rate = 1  # You can adjust this value based on your preference
    score = max(0, max_score - decrease_rate * exception_list_length)
    return score

def mean_arrival_score(expanded_relevant_flight):
    x = expanded_relevant_flight['Arrival Delay score']
    max_score = 100
    decrease_rate = 1
    score = max(0, max_score - decrease_rate * x.iloc[0])  # Adjust this line to fit your specific logic
    return score

def solution_ranking(The_flights, ex_lists):
    mean_arrivalscore = []
    less_exception_scores = []
    for i, exception_list in enumerate(ex_lists):
        score = mean_arrival_score(The_flights.iloc[[i]])
        mean_arrivalscore.append(score)
        exception_list_length = len(exception_list)
        less_exception_scores.append(calculate_score(exception_list_length))

    Rank = np.add(less_exception_scores, mean_arrivalscore)
    return Rank

def scores_passenger(passenger_df):
    passenger_df['Passenger'] = ''
    passenger_df['Cabin']=''
    passenger_df['Cabin_score'] = ''
    passenger_df['Class_score'] = ''
    passenger_df['Downline_score'] = ''
    passenger_df['PAX_score'] = 0
    passenger_df['SSR_score'] = 0
    passenger_df['Total Score'] = 0

    for index, row in passenger_df.iterrows():
        _class, downline, pax = row.COS_CD, row.SEG_SEQ, row.PAX_CNT

        cabin, cabin_score, class_scores, downline_score, pax_score, ssr_score, total_score = calculate_passenger_score(_class, downline, pax, 0)

        passenger_df.at[index, 'Passenger'] = f'Passenger {index + 1}'
        passenger_df.at[index,'Cabin']=cabin
        passenger_df.at[index, 'Cabin_score'] = cabin_score
        passenger_df.at[index, 'Class_score'] = class_scores
        passenger_df.at[index, 'Downline_score'] = downline_score
        passenger_df.at[index, 'PAX_score'] = pax_score
        passenger_df.at[index, 'SSR_score'] = ssr_score
        passenger_df.at[index, 'Total Score'] = total_score

    return passenger_df

def solution_inventory(acc_inventory, relevant_flights):
    relevant_rows = acc_inventory[
        (acc_inventory['ScheduleId'].isin(relevant_flights['ScheduleID'])) &
        (pd.to_datetime(acc_inventory['DepartureDate']).isin(pd.to_datetime(relevant_flights['Date'])))
    ].reset_index(drop=True)

    for i in range(len(relevant_rows)):
        relevant_rows.at[i, 'FC_CD_a'] = ''
        relevant_rows.at[i, 'BC_CD_a'] = ''
        relevant_rows.at[i, 'PC_CD_a'] = ''
        relevant_rows.at[i, 'EC_CD_a'] = ''

        relevant_rows.at[i, 'FC_CD_a'] = allocate_seats(relevant_rows['FC_Oversold'][i], relevant_rows['FC_CD'][i])
        relevant_rows.at[i, 'BC_CD_a'] = distribute_equally(relevant_rows['BC_Oversold'][i], relevant_rows['BC_CD'][i])
        relevant_rows.at[i, 'PC_CD_a'] = distribute_equally(relevant_rows['PC_Oversold'][i], relevant_rows['PC_CD'][i])
        relevant_rows.at[i, 'EC_CD_a'] = distribute_equally(relevant_rows['EC_Oversold'][i], relevant_rows['EC_CD'][i])

    return relevant_rows
passenger_original_df = pd.read_csv('A:\OneDrive - iittp.ac.in\IIT Tirupati\Activities\Inter IIT\Dataset\PNRB-ZZ-20231208_062017.csv')
passenger_individual = pd.read_csv('A:\OneDrive - iittp.ac.in\IIT Tirupati\Activities\Inter IIT\Dataset\PNRP-ZZ-20231208_111136.csv')

for i in range(len(passenger_original_df)):
    if passenger_original_df['DEP_KEY'][i]==selected_DEP_KEY:
        selected_DepartureAirport=passenger_original_df['ORIG_CD'][i]
        selected_ArrivalAirport=passenger_original_df['DEST_CD'][i]
        selected_dates=passenger_original_df['DEP_DT'][i]
        dt_object = passenger_original_df['DEP_DTML'][i]
        selected_DepartureTime = dt_object[-5:]
        break
    

passenger_origin=passenger_original_df[passenger_original_df['DEP_KEY']==selected_DEP_KEY]


# Merge the dataframes based on the RECLOC column
passenger_individual_data = pd.merge(passenger_individual, passenger_origin, on='RECLOC')
passenger_data = scores_passenger(passenger_individual_data)        # Assign scores to passengers


## Section 2: Find Relevant Flights
  # Create a DataFrame to store results
def score_flights(selected_date, flight_df, selected_ScheduleID):
    result_df = pd.DataFrame(columns=['ScheduleID', 'DepartureTime', 'Date', 'Date_departure_difference', 'Date_arrival_difference'])
    flight_df['DepartureTime'] = pd.to_datetime(flight_df['DepartureTime'], format='%H:%M')
    flight_df['ArrivalTime'] = pd.to_datetime(flight_df['ArrivalTime'], format='%H:%M')

    for i, dates_str in enumerate(flight_df['DepartureDates'].apply(lambda x: ''.join(char for char in x if char.isdigit() or char == '/' or char.isspace()).split())):
        for date_str in dates_str:
            date_in_array = pd.to_datetime(date_str, format='%m/%d/%Y')
            selected_datetime = pd.to_datetime(selected_date, format='%m/%d/%Y')
            selected_departure_result = (selected_datetime.replace(hour=flight_df.at[i, 'DepartureTime'].hour, minute=flight_df.at[i, 'DepartureTime'].minute)).strftime('%m/%d/%Y %H:%M')
            selected_arrival_result = (selected_datetime.replace(hour=flight_df.at[i, 'ArrivalTime'].hour, minute=flight_df.at[i, 'ArrivalTime'].minute)).strftime('%m/%d/%Y %H:%M')

            date_departure_difference = -(pd.to_datetime(selected_departure_result, format='%m/%d/%Y %H:%M') - date_in_array.replace(hour=flight_df.at[i, 'DepartureTime'].hour, minute=flight_df.at[i, 'DepartureTime'].minute))
            date_arrival_difference = -(pd.to_datetime(selected_arrival_result, format='%m/%d/%Y %H:%M') - date_in_array.replace(hour=flight_df.at[i, 'ArrivalTime'].hour, minute=flight_df.at[i, 'ArrivalTime'].minute))

            if 0 < date_departure_difference.days < 3 and flight_df['DepartureAirport'][i].strip()[0:3]==selected_DepartureAirport and flight_df['ArrivalAirport'][i].strip()[0:3]==selected_ArrivalAirport:
                result_df = pd.concat([result_df, pd.DataFrame({'ScheduleID': [flight_df.at[i, 'ScheduleID']],
                                                                'DepartureTime': [flight_df.at[i, 'DepartureTime']],
                                                                'Date': [date_str],
                                                                'Date_departure_difference': [date_departure_difference],
                                                                'Date_arrival_difference': [date_arrival_difference]})], ignore_index=True)

    expanded_relevant_flights = pd.DataFrame(columns=flight_df.columns)

    for _, row in flight_df[flight_df['ScheduleID'].isin(result_df['ScheduleID'])].iterrows():
        corresponding_dates = result_df.loc[result_df['ScheduleID'] == row['ScheduleID'], ['Date', 'Date_arrival_difference']]

        for date, date_diff in corresponding_dates.itertuples(index=False):
            new_row = row.copy()
            new_row['Date'] = date
            new_row['Date_arrival_difference'] = date_diff.total_seconds() / 3600
            new_row['Date_departure_difference'] = date_diff.total_seconds() / 3600
            expanded_relevant_flights = pd.concat([expanded_relevant_flights, pd.DataFrame([new_row])], ignore_index=True)

    expanded_relevant_flights.drop(['Frequency', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday', 'NoOfDepartures', 'Frequency_per_week'], axis=1, inplace=True)

    
    for _, row in expanded_relevant_flights.iterrows():
        arrival_delay, spf = row['Date_arrival_difference'], row['Date_departure_difference']

        # Directly subtract Timestamps
        CP_waiting_time_difference = (row['ArrivalTime'] - row['DepartureTime']).total_seconds() / 3600

        CP_same_airport, Equipment = row['DepartureAirport'] == row['ArrivalAirport'], row['ScheduleID'] == selected_ScheduleID

        arrival_delay_score, spf_score, city_pairs_score, equipment_score, total_flight_score = calculate_flight_score(arrival_delay, spf, Equipment, CP_waiting_time_difference, CP_same_airport)
        grade = assign_grade(total_flight_score)

        expanded_relevant_flights.loc[_, ['Arrival Delay score', 'SPF score', 'City pairs score', 'Equipment score', 'Total Score', 'Grade']] = arrival_delay_score, spf_score, city_pairs_score, equipment_score, total_flight_score, grade

    return expanded_relevant_flights
flight_data = pd.read_csv('A:\OneDrive - iittp.ac.in\IIT Tirupati\Activities\Inter IIT\Dataset\SCH-ZZ-20231208_035117.csv')
for i in range(len(flight_data)):
    if flight_data['DepartureTime'][i]==selected_DepartureTime:
        selected_scheduleID=flight_data['ScheduleID'][i]
        break

Individual_flights = score_flights(selected_dates, flight_data, selected_scheduleID)
Individual_flights
   #Solution Ranking
## Single impact flight=>Single alternate flight solutions
 # Read Inventory file from drive
inventory_df = pd.read_csv('A:\OneDrive - iittp.ac.in\IIT Tirupati\Activities\Inter IIT\Dataset\INV-ZZ-20231208_041852.csv')

  #Number of flights which have ETD<72 hours
Available_flights=Individual_flights
 # Convert text part into object part
Formattable_inventory=accessible_inventory(inventory_df)
 #Allocate available class seats to each class

Available_inventory=solution_inventory(Formattable_inventory,Available_flights)

 # Compute passenger and exception list data for each alternate solution
#passenger_names=passenger_data

inventory=Available_inventory
passenger_df = passenger_data.sort_values(by='Total Score', ascending=False)
passenger_df['Proposed_Class'] = ''
passenger_lists = []
exception_lists = []
passenger_copy_df = passenger_df.copy()

for i in range(len(inventory)):
        passenger_df = passenger_copy_df
        exception_list = pd.DataFrame(columns=passenger_df.columns)
        
        for _, passenger in passenger_df.iterrows():
            passenger_class = passenger['Cabin']
            classes = inventory[['FC_CD_a', 'BC_CD_a', 'PC_CD_a', 'EC_CD_a']].iloc[[i]].to_dict('records')[0]
            for class_key, class_seats in classes.items():
                proposed_class = next((pc for pc in [passenger_class] + proposed_classes.get(passenger_class, []) if class_seats.get(pc, 0) > 0), None)
                if proposed_class:
                    inventory.at[i, class_key][proposed_class] -= 1
                    passenger_df.at[passenger.name, 'Proposed_Class'] = proposed_class
                    break
            else:
                exception_list = pd.concat([exception_list, passenger_df.loc[[passenger.name]].drop(columns=['Proposed_Class'])], ignore_index=True)

        current_exception_list = exception_list
        current_passenger_df = passenger_copy_df[~passenger_copy_df['RECLOC'].isin(current_exception_list['RECLOC'])]

  # Move these lines outside the inner loop
        passenger_lists.append(current_passenger_df)
        exception_lists.append(current_exception_list)


#[passenger_lists, exception_lists] = solution_file(passenger_data, Available_inventory)
 # Rank solutions

Solution_ranks=solution_ranking(Available_flights,exception_lists)
Available_flights['Exception']=''
for i in range(len(exception_lists)):
  Available_flights['Exception'][i]=len(exception_lists[i])
Available_flights['Rank']=Solution_ranks
Available_flights = Available_flights.sort_values(by='Rank', ascending=False)
Something=Individual_flights
for i in range(len(Available_flights)):
    x=Available_flights['DepartureTime'][i]
    dt_str = x.strftime("%Y-%m-%d %H:%M:%S")
    Available_flights['DepartureTime'][i]=dt_str[-8:-3]
    x=Available_flights['ArrivalTime'][i]
    dt_str = x.strftime("%Y-%m-%d %H:%M:%S")
    Available_flights['ArrivalTime'][i]=dt_str[-8:-3]
 # Convert text part into object part

Available_inventory=solution_inventory(inventory_df,Something)
# Printing solutions to excel file
with pd.ExcelWriter('A:\OneDrive - iittp.ac.in\IIT Tirupati\Activities\Inter IIT\Dataset\Solution_file.xlsx', engine='xlsxwriter') as writer:
  Available_flights.to_excel(writer, sheet_name='Flights', index=False)
  Available_inventory.to_excel(writer, sheet_name='Available_seats', index=False)
  for i, passenger_list in enumerate(passenger_lists):
    passenger_list.to_excel(writer, sheet_name=f"Passenger List {i + 1}", index=False)

  for i, exception_list in enumerate(exception_lists):
    exception_list.to_excel(writer, sheet_name=f"Exception List {i + 1}", index=False)




```
